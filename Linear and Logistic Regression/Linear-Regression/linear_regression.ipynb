```python
import numpy as np
import matplotlib.pyplot as plt

# Set the seed for reproducibility
np.random.seed(42)

# Number of samples
num_samples = 100

# Generate random x values (features)
x_train = 2 * np.random.rand(num_samples, 1)

# Generate corresponding y values with a linear relationship (y = 4 + 3x + noise)
true_slope = 3
true_intercept = 4
noise = np.random.randn(num_samples, 1)

y_train = true_intercept + true_slope * x_train + noise

# Plotting the data
plt.figure(figsize=(8, 5))
plt.scatter(x_train, y_train, color="blue", label="Training data", alpha=0.7)
plt.title("Dummy Linear Regression Data")
plt.xlabel("x_train")
plt.ylabel("y_train")
plt.legend()
plt.grid(True)
plt.show()

print("Features shape is:",x_train.shape)
print("Target shape is:",y_train.shape)
```

```python
epochs = 50
b = 3
w = 4
j_wb_ = []
learning_rate = 0.01
m = len(x_train)
for epoch in range(epochs):
    j_wb = 0
    dw = 0
    db =0
    for i in range(m):
        yp = w * x_train[i] + b

        # Cost calculation
        j_wb =  j_wb + (yp - y_train[i]) ** 2

        # Gradients
        dw = dw + (x_train[i] * (yp - y_train[i]))
        db = db + (yp - y_train[i])
        
    # Average cost
    j_wb = (1 / ( 2 * m)) * j_wb
    j_wb_.append(j_wb)
    
    dw = 1/m * dw
    db = 1/m * db

    # Update parameters
    w = w - learning_rate * dw
    b = b - learning_rate * db

    if epoch % 10 == 0:
        print(f"Epoch {epoch}: Cost = {j_wb}, w = {w}, b = {b}")

print("\nFinal Cost:", j_wb)
print("Final Parameters w:", w, ", b:", b)
```

```python
for j in range(100):
    # Linear model prediction: y = w*x + b
    yp = w * x_train + b
    
# Print the shape of predicted output
print(yp.shape)
```

```python
plt.figure(figsize=(8, 5))
plt.scatter(x_train,y_train, color="blue", label="Training data", alpha=0.7)
plt.plot(x_train,yp,color="r", label="Prediction Line", linewidth=2 )
plt.title("Linear Regression Fit")
plt.xlabel("Input Feature (x)")
plt.ylabel("Target Value (y)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8, 5))
plt.plot(j_wb_, color="y", label="Cost Function")
plt.title("Training Curve")
plt.xlabel("Iterations")
plt.ylabel("Cost")
plt.legend()
plt.grid(True)
plt.show()
```
